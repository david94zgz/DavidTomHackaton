```python
import pandas
import json

df = pandas.read_csv("data/vgsales.csv")
```

```python
df['json'] = df.apply(lambda x:x.to_json(), axis=1)
#json_df = df['json']
#json_df.head()
```

```python
query = "Sports"
df_selection = df.loc[df["Genre"] == query]
result = {
    "num_entries": df_selection.size,
    "data": 
        [json.loads(x) for x in df_selection['json'].to_dict().values()]
}
print(result)
```
